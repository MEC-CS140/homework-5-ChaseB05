```python
num_list = [2, 7, 10, 100, 101, 200 ,270, 947]

# Task 1: Print the length of the list
print("Length of num_list:", len(num_list))  # Output: 8

# Task 2: Remove 100 from num_list
num_list.remove(100)
print("After removing 100:", num_list)  # Output: [2, 7, 10, 101, 200, 270, 947]

# Task 3: Check if 31 exists in the list
if 31 in num_list:
    print("31 exists in the list.")
else:
    print("31 does not exist in the list.")

# Task 4: Fizz Buzz
for i in range(1, 101):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)

# Task 5: Create a function that adds two numbers together and returns the sum
def add(num1, num2):
    return num1 + num2

# Task 6: Create a function that subtracts two numbers together and returns the difference
def subtract(num1, num2):
    return num1 - num2
```
